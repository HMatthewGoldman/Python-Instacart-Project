## 04 Question 5 Customer Profiling

### Age Profiling Variable

```python
df['Age'].value_counts()
```

```python
round(df['Age'].describe(),2)
```

```python
df['Age'].isnull().sum()
```

```python
age_group=[]
for value in df['Age']:
    if (value<=39):
        age_group.append('under 40')
    elif (value>=40)&(value<=60):
        age_group.append('40-60')
    elif (value>=61):
        age_group.append('over 60')
    else:
        age_group.append('No Age Group Assigned')
```

```python
age_group
```

```python
df['age_groups']=age_group
```

```python
df['age_groups'].value_counts()
```

### Income Profiling Variable

```python
df['income'].value_counts()
```

```python
round(df['income'].describe(),2)
```

```python
income_range=[]
for value in df['income']:
    if value<100000:
        income_range.append('under 100k')
    elif (value>=100000)&(value<=300000):
        income_range.append('100-300k')
    elif value>300000:
        income_range.append('over 300k')
    else:
        income_range.append('No Income Range Assigned')
```

```python
income_range
```

```python
df['income_range']=income_range
```

```python
df['income_range'].value_counts()
```

### Dependant Profiling Variable

```python
df['n_dependants'].value_counts()
```

```python
round(df['n_dependants'].describe(),2)
```

```python
parents = []
for value in df['n_dependants']:
    if value==0: 
        parents.append('Not Parent')
    elif value>0:
        parents.append('Parent')
    else:
        parents.append('No Dependants Assigned')
```

```python
parents
```

```python
df['parents']=parents
```

```python
df['parents'].value_counts()
```

### Create profile based on age and dependants

```python
df.loc[(df['age_groups']=='under 40')&(df['parents']=='Parent'), 'age_dep_profile']='Young Parent'
```

```python
df.loc[(df['age_groups']=='under 40')&(df['parents']=='Not Parent'), 'age_dep_profile']='Young Non Parent'
```

```python
df.loc[(df['age_groups']=='40-60')&(df['parents']=='Parent'), 'age_dep_profile']='Mid Age Parent'
```

```python
df.loc[(df['age_groups']=='40-60')&(df['parents']=='Not Parent'), 'age_dep_profile']='Mid Age Non Parent'
```

```python
df.loc[(df['age_groups']=='over 60')&(df['parents']=='Parent'), 'age_dep_profile']='Older Parent'
```

```python
df.loc[(df['age_groups']=='over 60')&(df['parents']=='Not Parent'), 'age_dep_profile']='Older Non Parent'
```
